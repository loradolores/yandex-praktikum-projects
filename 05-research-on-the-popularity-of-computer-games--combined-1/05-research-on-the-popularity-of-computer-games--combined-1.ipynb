### 4.2. Типы данных и пропуски

````python
games['year_of_release'].value_counts()
````

В столбце `year_of_release` содержится всего 269 пропусков, при общем количестве строк в датафрейме - 16715. Поэтому эти 269 строк с пропусками можно удалить.
````python
games = games.dropna(subset=['year_of_release']).reset_index(drop=True)
games['year_of_release'].isna().sum()
````

````python
games.info()
````

Посмотрим количество пропусков в столбце `user_score`.

````python
games['user_score'].isna().sum()
````

Также в столбце `user_score` есть строковые данные "tbd" в количестве 2376.

````python
games['user_score'].value_counts()
````

В столбце `user_score` содержится очень много пропусков - 6607, при общем количестве строк в датафрейме - 16715. Это можно объяснить тем, что пользователи не всегда оставляют оценки продукту, который они покупают. Такое большое количество строк с  пропусками нельзя удалять, иначе мы потеряем огромную часть датафрейма. Из-за содержащихся в нём строковых значений "tbd", столбец имеет тип `object`. Для этого необходимо заменить значения "tbd" в столбце `user_score` на значения "NaN" (имеющие тип `float`), т.к. мы не можем вписать конкретные значения рейтинга, чтобы не исказить данные. #### 5.5.1. Влияние отзывов на продажи платформы "PS4"

Рассмотрим популярную платформу "PS4".

````python
# отзывы критиков
games[(games['platform'] == 'PS4') & (games['year_of_release'] >= 2014)]\
.plot(x='critic_score', y='total_sales', figsize=(8, 6), kind='scatter', 
      title='Отзывы критиков о платформе PS4')
plt.xlabel('Оценка критиков (максимум 100)')
plt.ylabel('Суммарные продажи (миллионы проданных копий)')
plt.show()

critic_ratio = games[(games['platform'] == 'PS4')\
                     & (games['year_of_release'] >= 2014)]['critic_score']\
.corr(games['total_sales'])

print(f'Коэффициент корреляции Пирсона: {critic_ratio:.2f}')


# отзывы пользователей
games[(games['platform'] == 'PS4') & (games['year_of_release'] >= 2014)]\
.plot(x='user_score', y='total_sales', figsize=(8, 6), kind='scatter', 
      title='Отзывы пользователей о платформе PS4')
plt.xlabel('Оценка пользователей (максимум 10)')
plt.ylabel('Суммарные продажи (миллионы проданных копий)')
plt.show()

user_ratio = games[(games['platform'] == 'PS4')\
                   & (games['year_of_release'] >= 2014)]['user_score']\
.corr(games['total_sales'])

print(f'Коэффициент корреляции Пирсона: {user_ratio:.2f}')
````

### 7.1. Гипотеза 4

Средние пользовательские рейтинги платформ "Xbox One" и "PC" за период c 2014 по 2016 гг. одинаковые.

```
H_0: Средний пользовательский рейтинг (`user_score`) платформы "Xbox One" = средний пользовательский рейтинг (`user_score`) платформы "PC".
H_a: Средний пользовательский рейтинг (`user_score`) платформы "Xbox One" ≠ средний пользовательский рейтинг (`user_score`) платформы "PC".
alpha = 0.05
```

Удалим пропуски в столбце `user_score`.

````python
games_user_score = actual_games.dropna(subset=['user_score'])\
.reset_index(drop=True)

games_user_score['user_score'].isna().sum()
````

Проверим уникальные значения в столбце `platform` (платформа "Xbox One" внесена под названием "XOne").

````python
games_user_score['platform'].unique()
````

````python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    games_user_score[games_user_score['platform'] == 'XOne']['user_score'], 
    games_user_score[games_user_score['platform'] == 'PC']['user_score'],
    equal_var = False)

# alpha = задайте значение уровня значимости
alpha = .05

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

